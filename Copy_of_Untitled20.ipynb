```
import numpy as np
import tensorflow as tf
import base64
import IPython.display as display
import io
import PIL.Image as Image
xc=["train00-2071.tfrec","train01-2071.tfrec","train02-2071.tfrec","train03-2071.tfrec"]
def _parse_image_function(example_proto):
  # Parse the input tf.Example proto using the dictionary above.
    return tf.io.parse_single_example(example_proto, image_feature_description)
c=0
for g in xc:
  raw_image_dataset = tf.data.TFRecordDataset("./content/unzip/"+g)
# Create a dictionary describing the features.
  image_feature_description = {
    'image_name': tf.io.FixedLenFeature([], tf.string),
    'image': tf.io.FixedLenFeature([], tf.string)
    
  }
  parsed_image_dataset = raw_image_dataset.map(_parse_image_function)
  parsed_image_dataset
  for raw_record in parsed_image_dataset:
    image_raw = raw_record['image'].numpy()
    io1=Image.open(io.BytesIO(image_raw))
    io1.save("./content/"+str(((raw_record['image_name'].numpy()))).strip("b")+".jpg")
    io1.close()
    image_raw=None
    c+=1
```

```
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation,Flatten
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Conv2D, GlobalMaxPooling1D,AveragePooling2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Activation, Dense
model = Sequential()
model.add(Conv2D(28, (3, 3), padding="same",input_shape=(4000,6000,3)))
model.add(Activation("softmax"))
model.add(MaxPooling2D(pool_size=(2,2),strides=(1,1),padding='same'))
model.add(Activation("softmax"))
model.add(Flatten())
model.add(Dropout(0.20))
model.add(Dense(10))
model.add(Activation("softmax"))
model.compile(loss='binary_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```
