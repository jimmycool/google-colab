```python
import numpy as np
import matplotlib.pyplot as plt
X=[0.5,2.5]
Y=[0.2,0.9]
x=np.array(X)
y=np.array(Y)
plt.plot(x,y)
plt.show()
```

```python
def sigmoid(x,w,b):
  return (1/(1+np.exp(-w*x-b)))
y1=sigmoid(x[0],0.5,0)
y2=sigmoid(x[1],0.5,0)
loss1=(y[0]-y1)**2
loss2=(y[1]-y2)**2
print("total loss=",(loss1+loss2)/2)
```

```python
w=np.array([0.5,-0.5,0.84,1.36,1.60,1.75])
b=np.array([0,0,-0.73,-1.56,-2,-2.25])
table1=[]
table2=[]
table3=[]
for t in range(0,len(w)):
  y1=sigmoid(x[0],w[t],b[t])
  y2=sigmoid(x[1],w[t],b[t])
  loss1=(y[0]-y1)**2
  loss2=(y[1]-y2)**2
  table1.append(loss1)
  table2.append(loss2)
  table3.append((loss1+loss2)/2)
print("x"+"\t\t\t"+"y"+"\t\t\t\t"+"loss")  
for t in range(0,len(table1),1):  
  print(table1[t],table2[t],table3[t])  
```

```python
y5=sigmoid(x[0],w[len(w)-1],b[len(b)-1])
y6=sigmoid(x[1],w[len(w)-1],b[len(b)-1])
print(y5,y6)
plt.plot([y5,y6])
plt.show()
```
